```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Show color image in matplotlib
img_color = cv2.imread('../image/messi5.jpg')
#opencv use BGR, but matplotlib use RGB, need to convert color scheme
RGB_img = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```
