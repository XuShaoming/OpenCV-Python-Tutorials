### 1.3.3 Performance Measurement and Improvement Techniques

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
e1 = cv2.getTickCount()
# your code execution
e2 = cv2.getTickCount()
time = (e2 - e1) / cv2.getTickFrequency()
print(time)
```

```python
img1 = cv2.imread('messi5.jpg')
e1 = cv2.getTickCount()
for i in range(5,49,2):
    img1 = cv2.medianBlur(img1,i)
e2 = cv2.getTickCount()
t = (e2 - e1)/cv2.getTickFrequency()
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
print(t)
```
