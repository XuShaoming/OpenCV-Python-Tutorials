```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
img = cv2.imread('../image/messi5.jpg')
##Accessing and Modifying pixel values
print("Accessing and Modifying pixel values")
px = img[100,100]
print(px)
blue = img[100,100,0]
print(blue)
img[100,100] = [255,255,255]
print(img[100,100])

#Better method:
print("\nBetter method")
print(img.item(10,10,2))
img.itemset((10,10,2),100)
print(img.item(10,10,2))

##Accessing Image Properties
print("\n\nAccessing Image Properties")
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
#A region of interest (ROI) for image
img_origin = cv2.imread('../image/messi5.jpg')
img = cv2.imread('../image/messi5.jpg')
ball = img[280:340, 330:390]
img[273:333, 100:160] = ball
img_origin = cv2.cvtColor(img_origin, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


plt.subplot(121), plt.imshow(img_origin), plt.title("origin")
plt.subplot(122), plt.imshow(img), plt.title("ROI")
plt.show()
```

```python
##Splitting and Merging Image Channels
img = cv2.imread('../image/messi5.jpg')
#a new way to convert bgr to rgb
b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))
img_r0 = cv2.merge((r,g,b))

# make all the red to zero
img_r0[:,:,0] = 0


plt.subplot(121), plt.imshow(img), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_r0),plt.title('red to zero')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Making Borders for Images (Padding)

BLUE = [255,0,0]
img1 = cv2.imread('../image/opencv_logo.png')
replicate = cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_CONSTANT,value=BLUE)


plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
