## 1.4.3 Geometric Transformations of Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib inline
```

### Scaling

```python
img = cv2.imread('../image/messi5.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Translation

```python
img = cv2.imread('../image/messi5.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(1,2,1), plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.subplot(1,2,2), plt.imshow(dst,cmap = 'gray', interpolation = 'bicubic')
```

### Rotation

```python
img = cv2.imread('../image/messi5.jpg',0)
rows,cols = img.shape
#cv2.getRotationMatrix2D(center, angle, scale)  retval
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
print(M)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap='gray', interpolation='bicubic')
```

### Affine Transformation

```python
#cv2.getAffineTransform(src, dst)  retval
img = cv2.imread('../image/drawing.png')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
print(M)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

### Perspective Transformation

```python
img = cv2.imread('../image/sudokusmall.png')
rows,cols,ch = img.shape
pts1 = np.float32([[65,65],[130,65],[65,130],[130,130]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(200,200))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
